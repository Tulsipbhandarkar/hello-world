```
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv("iris1 (1).csv")
df
```

```
df=df.drop('sepal_length',axis='columns')
df=df.drop('sepal_width',axis='columns')
df
```

```
from sklearn.cluster import KMeans 
sse=[]
k_rng=range(1,10)
for k in k_rng:
  km=KMeans(n_clusters=k)
  km.fit(df)
  sse.append(km.inertia_ )
plt.xlabel('K')
plt.ylabel('sum of squared error')
plt.plot(k_rng,sse)
```

```
km=KMeans(n_clusters=3)
km.fit_transform(df)
y_pred=km.fit_predict(df)
df['cluster']=y_pred
df0=df[df.cluster==0]
df1=df[df.cluster==1]
df2=df[df.cluster==2]
plt.scatter(df0.petal_length,df0.petal_width,color="green")
plt.scatter(df1.petal_length,df1.petal_width,color="yellow")
plt.scatter(df2.petal_length,df2.petal_width,color='red')
plt.scatter(km.cluster_centers_[:,0],km.cluster_centers_[:,1],s=100,color='blue',marker='*',label='centroid')
plt.legend()
km.cluster_centers_
```
